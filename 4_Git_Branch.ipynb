# What is a Git Branch?

- A branch in Git = a separate line of development.

- It allows you to work on new features or fixes without affecting the main project (main branch).

- Once the work is ready, you can merge it back to the main branch.

<h4>Analogy</h4>

- Main branch = the main road

- Feature branch = a side road where you try something new

- Merge = join the side road back to the main road

# Why Use Branches?

- Experiment safely without breaking the main code.

- Multiple people can work on different features at the same time.

- Keep the main branch stable for production.

<h5>View Branches</h5>

- git branch

<h5>Create a New Branch</h5>

<h5>Switch to a Branch</h5>

- git switch feature1

<h5>Create and Switch in One Step</h5>

- git switch -c feature2

<h5>Merge a Branch into Main</h5>

- git switch main
- git merge feature1
- #Combines changes from feature1 into main

<h5>Delete a Branch</h5>

- git branch -d feature1
- #Deletes a branch locally after it is merged.
- #For remote branch:
- git push origin --delete feature1

## Branching Best Practices

- Always create a new branch for each feature or bug fix.
- Keep main branch stable.
- Give branches clear names, e.g., feature/login, bugfix/header.
